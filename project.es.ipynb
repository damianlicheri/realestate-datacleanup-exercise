#### Ejercicio 05. ¿El data set contiene valores no admitidos (NAs)? (★☆☆)

Imprima un booleano (`true` o `fase`) seguido de la fila/columna que contiene el NAs.

```python
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')

na_rows = []
na_columns = []
for i in range(ds.shape[0]):
    for j in range(ds.shape[1]):
        if pd.isnull(ds.iloc[i, j]):
            na_rows.append(i)
            na_columns.append(ds.columns[j])

if na_rows:
    print(f"Se encontraron valores NA en las siguientes filas y columnas:")
    for i, j in zip(na_rows, na_columns):
        print(f"Fila {i+1}, Columna {j}")
else:
    print("No se encontraron valores NA en el archivo.")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

```python
import pandas as pd
import matplotlib.pyplot as plt

ds = pd.read_csv("assets/real_estate.csv", sep=";")

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_subset = ds[ds["level5"].isin(poblaciones)]

ds_subset["price_norm"] = ds_subset["price"] / ds_subset["price"].mean()

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

for i, poblacion in enumerate(poblaciones):
    ax = axes[i // 2, i % 2]
    ds_subset[ds_subset["level5"] == poblacion].plot.hist(ax=ax, x="price_norm", 
                                                         title=poblacion, legend=False)

plt.tight_layout()
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in the question 14 (★★☆)

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
import pandas as pd

ds = pd.read_csv("assets/real_estate.csv", sep=";")

poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
ds_subset = ds[ds["level5"].isin(poblaciones)]

getafe = ds_subset.loc[ds_subset["level5"] == "Getafe"]
alcorcon = ds_subset.loc[ds_subset["level5"] == "Alcorcón"]

getafe["precio_m2"] = getafe["price"] / getafe["surface"]
alcorcon["precio_m2"] = alcorcon["price"] / alcorcon["surface"]

print("Getafe:")
print(f"Media: {getafe['precio_m2'].mean():.2f} €/m²")
print(f"Mediana: {getafe['precio_m2'].median():.2f} €/m²")
print(f"Desviación estándar: {getafe['precio_m2'].std():.2f} €/m²")

print("Alcorcón:")
print(f"Media: {alcorcon['precio_m2'].mean():.2f} €/m²")
print(f"Mediana: {alcorcon['precio_m2'].median():.2f} €/m²")
print(f"Desviación estándar: {alcorcon['precio_m2'].std():.2f} €/m²")

```

Alcorcón tiene un precio medio por metro cuadraro ligeramente más elevado que Getafe
